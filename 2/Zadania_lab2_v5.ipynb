## **Zadanie 1 - poziom podstawowy**
Wczytywanie konfiguracji z YAML

Napisz program konsolowy, który wczytuje konfigurację z pliku YAML.

### Założenia

**Program powinien:**

- przyjmować z linii poleceń ścieżkę do pliku (np. `--config config.yaml`),
- korzystać z argparse,
- wczytać plik YAML,
- zapisać konfigurację do słownika (`dict`),
- wyświetlić konfigurację (może być zwykłe `print(config)`),
- sprawdzić, czy plik istnieje,
- jeśli plik nie istnieje — wyświetlić komunikat błędu.

**Wymagania użycia:**
- argparse,
- pathlib.Path,
- biblioteka YAML (np.`pyyaml`).

**Efekt końcowy:**

Program:

- parsuje argument,
- wczytuje YAML,
- zapisuje dane do `dict`,
- wyświetla konfigurację,
- obsługuje brak pliku.

```yaml
# Przykładowy plik YAML
# Plik powinien zawierać zagnieżdżoną konfigurację:
app:
  name: "MojaAplikacja"
  debug: true

server:
  host: "127.0.0.1"
  port: 8080
  timeout: 30

database:
  credentials:
    user: "admin"
    password: "secret"
  settings:
    pool_size: 5
    retry: true

# Uwaga:
# Konfiguracja zawiera zagnieżdżone słowniki, co oznacza, że:
# nie wystarczy prosty print(dict), należy sprawdzić strukturę danych i zaprezentować ją czytelnie.
```

````python
# Przykład parsowania argumentów (argparse + Path)
# Program powinien korzystać z pathlib.Path jako typu argumentu:
import argparse
from pathlib import Path

parser = argparse.ArgumentParser(description="Wczytywanie konfiguracji YAML")

parser.add_argument(
    "--config",
    type=Path,
    required=True,
    help="Ścieżka do pliku konfiguracyjnego YAML"
)

args = parser.parse_args()

config_path = args.config
````

### Wskazówki
- Sprawdzenie pliku:

```python
if not config_path.exists():
    print("Plik nie istnieje!")
```

- Do czytelnego wyświetlania możesz:
  1. użyć pprint,
  2. albo napisać własną funkcję rekurencyjną do wypisywania zagnieżdżonych słowników.
