## **Teoria: Programowanie obiektowe w Pythonie**

# Wprowadzenie do programowania obiektowego

Programowanie obiektowe (OOP – Object-Oriented Programming) to paradygmat, w którym program budowany jest jako zbiór współpracujących obiektów. Każdy obiekt reprezentuje pewien fragment rzeczywistości i łączy dane (atrybuty) z operacjami na tych danych (metody).

Główne cele OOP:

* uporządkowanie kodu,
* zwiększenie czytelności i modularności,
* łatwiejsze utrzymanie i rozwój aplikacji,
* możliwość ponownego wykorzystania kodu.

# Klasy i obiekty

Klasa to szablon (definicja), na podstawie którego tworzone są obiekty.
Obiekt to konkretna instancja klasy.
link: https://docs.python.org/3/tutorial/classes.html

Przykładowo:

* klasa: Student,
* obiekt: konkretny student z danymi.

Klasa definiuje:

* jakie dane posiada obiekt (atrybuty),
* jakie operacje można na nim wykonać (metody).

# Atrybuty i metody

Atrybuty to zmienne przechowujące dane obiektu, np. imię, saldo.
Metody to funkcje zdefiniowane w klasie, które operują na tych danych.

Metody zawsze przyjmują jako pierwszy parametr odniesienie do obiektu (self).

# Konstruktor __init__() i słowo kluczowe self

Konstruktor `__init__()` jest specjalną metodą wywoływaną podczas tworzenia obiektu. Służy do inicjalizacji jego stanu.

`self` oznacza konkretny obiekt, na którym wykonywana jest metoda. Dzięki temu można odwoływać się do jego atrybutów i metod.

# Metody specjalne

Python udostępnia tzw. metody specjalne (dunder methods), które pozwalają definiować zachowanie obiektów - metody o nazwach otoczonych podwójnymi podkreśleniami, np. __str__, __repr__, __len__ . Służą one do definiowania, w jaki sposób obiekty własnych klas mają zachowywać się w różnych sytuacjach, np. podczas wyświetlania, porównywania, obliczania długości czy wykonywania operacji arytmetycznych.
link: https://docs.python.org/3/reference/datamodel.html#special-method-names

Metody te są wywoływane automatycznie przez interpreter lub funkcje wbudowane. Oznacza to, że programista zwykle nie uruchamia ich bezpośrednio, lecz korzysta z takich instrukcji jak print(obiekt), repr(obiekt) albo len(obiekt), a Python sam odwołuje się wtedy do odpowiedniej metody specjalnej. Dzięki temu obiekty klas użytkownika mogą działać podobnie do typów wbudowanych, takich jak napisy, listy czy słowniki.
Najważniejsze:

`__str__()` – zwraca czytelną reprezentację tekstową obiektu, przeznaczoną głównie dla użytkownika,
`__repr__()` – zwraca bardziej techniczną reprezentację obiektu, przydatną dla programisty i podczas debugowania,
`__len__()` – pozwala określić, jaką długość ma obiekt, dzięki czemu można użyć funkcji len().

W praktyce oznacza to, że:

* print(obiekt) korzysta zwykle z __str__(),
* repr(obiekt) korzysta z __repr__(),
* len(obiekt) korzysta z __len__()

Zastosowanie metod specjalnych zwiększa czytelność i funkcjonalność programu. Pozwalają one tworzyć klasy, których obiekty zachowują się w sposób naturalny i intuicyjny.

# Enkapsulacja

Enkapsulacja polega na ukrywaniu szczegółów implementacji i kontrolowaniu dostępu do danych.


W Pythonie wyróżnia się trzy poziomy dostępu do atrybutów:

* atrybuty publiczne – dostępne bez ograniczeń, np. `obiekt.nazwa`,
* atrybuty chronione – zapisywane zwykle jako `_nazwa`; zgodnie z konwencją oznaczają, że nie powinny być używane bezpośrednio poza klasą lub jej podklasami,
* atrybuty prywatne – zapisywane jako `__nazwa`; Python utrudnia do nich bezpośredni dostęp spoza klasy.

Dobrym przykładem jest klasa KontoBankowe. Gdyby saldo było zwykłym publicznym atrybutem, można byłoby ustawić je na dowolną wartość, np. ujemną albo przypadkowo je nadpisać. Zastosowanie enkapsulacji pozwala ukryć saldo wewnątrz klasy i udostępnić jedynie metody takie jak wplac() i wyplac(). Dzięki temu można sprawdzić, czy wpłacana kwota jest dodatnia i czy użytkownik nie próbuje wypłacić więcej pieniędzy, niż ma na koncie.

Korzyści:

* bezpieczeństwo danych,
* kontrola poprawności operacji,
* lepsza organizacja kodu.

# Właściwości (@property)

Dekorator @property pozwala traktować metodę jak atrybut.
link: https://docs.python.org/3/library/functions.html#property

Umożliwia:

* kontrolowany odczyt danych,
* walidację przy ustawianiu wartości (setter),
* ukrycie implementacji.

Przykład zastosowania: bezpośredni dostęp do salda, ale bez możliwości jego niekontrolowanej modyfikacji.

# Relacje między obiektami

Obiekty mogą ze sobą współpracować. Jedna klasa może zawierać obiekty innej klasy.

Typowe relacje:

* kompozycja – obiekt składa się z innych obiektów (np. samochód zawiera silnik),
* agregacja – obiekty są powiązane, ale mogą istnieć niezależnie.

Pozwala to modelować rzeczywiste systemy.

# Dziedziczenie

Dziedziczenie pozwala tworzyć nowe klasy na podstawie już istniejących.
link: https://docs.python.org/3/tutorial/classes.html#inheritance

Klasa potomna:

* dziedziczy atrybuty i metody klasy bazowej,
* może je rozszerzać lub modyfikować.

**Przykład**

Modelujemy pojazdy:

* mamy klasę ogólną Pojazd,
* oraz bardziej szczegółowe: Samochod i Motocykl.


````python
class Pojazd:  # klasa bazowa (ogólna dla wszystkich pojazdów)

    def __init__(self, marka, model):  # konstruktor klasy
        self.marka = marka  # przypisanie atrybutu marka do obiektu
        self.model = model  # przypisanie atrybutu model do obiektu

    def opis(self):  # metoda zwracająca opis pojazdu
        return f"Pojazd: {self.marka} {self.model}"  # wykorzystanie atrybutów obiektu


class Samochod(Pojazd):  # klasa potomna dziedzicząca po Pojazd

    def __init__(self, marka, model, drzwi):  # konstruktor rozszerzony o nowy atrybut
        super().__init__(marka, model)  # wywołanie konstruktora klasy bazowej (inicjalizacja marka, model)
        self.drzwi = drzwi  # dodanie nowego atrybutu specyficznego dla samochodu

    def opis(self):  # nadpisanie metody z klasy bazowej (override)
        return f"Samochód: {self.marka} {self.model}, drzwi: {self.drzwi}"  # rozszerzony opis


class Motocykl(Pojazd):  # klasa potomna dziedzicząca po Pojazd

    def __init__(self, marka, model, typ):  # konstruktor rozszerzony o nowy atrybut
        super().__init__(marka, model)  # wywołanie konstruktora klasy bazowej
        self.typ = typ  # dodanie atrybutu specyficznego dla motocykla

    def opis(self):  # nadpisanie metody z klasy bazowej
        return f"Motocykl: {self.marka} {self.model}, typ: {self.typ}"  # inny sposób opisu obiektu
````


**Użycie**

```
s = Samochod("Toyota", "Corolla", 4)
m = Motocykl("Yamaha", "MT-07", "naked")

print(s.opis())
print(m.opis())
```

Korzyści:

* redukcja powtarzalności kodu,
* hierarchizacja struktur,
* łatwiejsze rozszerzanie systemu.


# Polimorfizm

Polimorfizm oznacza możliwość używania tych samych metod dla różnych obiektów, przy zachowaniu ich specyficznego działania.

Przykład:
różne klasy mogą mieć metodę `poruszaj_sie()`, ale każda implementuje ją inaczej.

Pozwala pisać bardziej elastyczny i uniwersalny kod.

# Nadpisywanie metod (override)

Klasa potomna może zmienić działanie metody odziedziczonej z klasy bazowej.

Pozwala to:

* dostosować zachowanie obiektu,
* rozszerzyć funkcjonalność,
* zachować wspólny interfejs.



# Metody klasowe i statyczne

Metody klasowe (@classmethod)

* operują na klasie, a nie na konkretnym obiekcie,
* często używane do tworzenia obiektów (np. fabryki).

Metody statyczne (@staticmethod)

* nie korzystają ani z obiektu, ani z klasy,
* pełnią funkcję pomocniczą (np. walidacja danych).

# Modelowanie problemów w OOP

Programowanie obiektowe służy do odwzorowania rzeczywistych problemów w kodzie.

Proces modelowania:

* identyfikacja obiektów (np. Student, Konto, Pojazd),
* określenie ich właściwości (atrybuty),
* określenie zachowań (metody),
* zdefiniowanie relacji między obiektami.

Dobrze zaprojektowany model:

* jest czytelny,
* łatwy do rozbudowy,
* odzwierciedla rzeczywistość.
